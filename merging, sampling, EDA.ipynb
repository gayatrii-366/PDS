```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url1 = "/content/sample_data/california_housing_test.csv"
url2 = "/content/sample_data/california_housing_train.csv"

dataset1 = pd.read_csv(url1)
dataset2 = pd.read_csv(url2)

print("\nDataset 1:")
print(dataset1.info())
print("\nMissing Values in Dataset 1:")
print(dataset1.isnull().sum())

print("\nDataset 2:")
print(dataset2.info())
print("\nMissing Values in Dataset 2:")
print(dataset2.isnull().sum())

dataset1_selected = dataset1.drop(['longitude', 'latitude', 'population',
                                   'housing_median_age', 'households',
                                   'median_income', 'median_house_value'], axis=1)
dataset2_selected = dataset2.drop(['longitude', 'latitude', 'population',
                                   'housing_median_age', 'households',
                                   'median_income', 'median_house_value'], axis=1)

print("\nDistribution of 'total_rooms' in Dataset 1:")
sns.histplot(dataset1['total_rooms'], kde=True, color='blue', label='Dataset 1')
plt.legend()
plt.show()

print("\nDistribution of 'total_rooms' in Dataset 2:")
sns.histplot(dataset2['total_rooms'], kde=True, color='orange', label='Dataset 2')
plt.legend()
plt.show()

merged_dataset = pd.concat([dataset1_selected, dataset2_selected], ignore_index=True)

print("\nEDA for Merged Dataset:")
print(merged_dataset.describe())
print("\nMissing Values in Merged Dataset:")
print(merged_dataset.isnull().sum())

print("\nCorrelation Heatmap for Merged Dataset:")
corr_matrix = merged_dataset.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

dataset1_sample = dataset1.sample(n=5, random_state=42)
dataset2_sample = dataset2.sample(n=5, random_state=42)

dataset1_sample = dataset1_sample.drop(['longitude', 'latitude', 'population',
                                        'housing_median_age', 'households',
                                        'median_income', 'median_house_value'], axis=1)
dataset2_sample = dataset2_sample.drop(['longitude', 'latitude', 'population',
                                        'housing_median_age', 'households',
                                        'median_income', 'median_house_value'], axis=1)

merged_sample = pd.concat([dataset1_sample, dataset2_sample], ignore_index=True)

print("\nEDA for Merged Sample:")
print(merged_sample.describe())
print("\nMissing Values in Merged Sample:")
print(merged_sample.isnull().sum())

print("\nDistribution of 'total_rooms' in Merged Sample:")
sns.histplot(merged_sample['total_rooms'], kde=True, color='green')
plt.title("Merged Sample - Distribution of 'total_rooms'")
plt.show()
```
